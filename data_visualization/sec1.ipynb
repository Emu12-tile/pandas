```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ibm=pd.read_csv("ibm.csv" ,parse_dates=["Date"],index_col="Date")
ibm.head()
```

```python
ibm.plot(y="Close")
ibm["Close"].plot()
```

```python
plt.style.available
```

```python
plt.style.use("grayscale")
ibm.plot(y="Close")
```

```python
plt.style.use("dark_background")
ibm.plot(y="Close")
```

```python
def rank_performance(stock_price):
    if stock_price<=50:
        return "Poor"
    elif stock_price > 50 and stock_price<=100:
        return "Satisfactory"
    else:
        return "Excellent"

ibm["Close"].apply(rank_performance).value_counts().plot(kind="pie", color="black")
```
